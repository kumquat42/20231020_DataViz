## corr matrix & heatmap

```python
corr_df = df.corr(numeric_only=True).round(2)
corr_df
```

```python
sns.heatmap(corr_df, annot=True, cmap='crest')
```

```python
upp_mat = np.triu(corr_df)

sns.heatmap(corr_df, annot=True, cmap='coolwarm', mask=upp_mat, center=0)
```

```python
# Top3 correlation
corr_df['SalePrice'].abs().sort_values(ascending=False)[0:4]
```
